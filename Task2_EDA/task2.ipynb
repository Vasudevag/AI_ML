```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")  # Same dataset from Task 1
df.head()
```

```python
df.describe()
df.info()
```

```python
df.hist(bins=20, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Boxplot of Age and Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived')
plt.show()
```

```python
df[['Age', 'Fare']].skew()
```

## Inferences:

- Fare and Age are right-skewed.
- Survived has correlation with Sex and Pclass.
- Outliers present in Fare.
- Embarked seems related to survival in some cases.
